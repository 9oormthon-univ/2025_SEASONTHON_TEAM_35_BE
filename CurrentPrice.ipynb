```python
import asyncio
from pykis import *
from dotenv import load_dotenv
import os 

# 발급받은 App Key, App Secret 설정
load_dotenv()
APP_KEY = os.getenv('APP_KEY')
APP_SECRET = os.getenv('APP_SECRET')

async def main():
    # 개인 계좌번호로 초기화
    kis = Pykis(
        appkey=APP_KEY,
        appsecret=APP_SECRET,
        virtual_account=False,  # 모의투자 계좌 사용 여부
    )

    # 실시간 시세 수신 콜백 함수
    def on_price(data):
        print(f"[{data.stock_code}] 현재가: {data.price}원, 등락률: {data.rate}%")

    # 삼성전자(005930) 실시간 시세 구독 시작
    await kis.ws_client.add_realtime_price("005930", on_price)

    # 10초간 실시간 시세 수신
    await asyncio.sleep(10)

    # 구독 중지
    await kis.ws_client.remove_realtime_price("005930")

    # 웹소켓 연결 종료
    await kis.ws_client.close()


if __name__ == "__main__":
    asyncio.run(main())
```

```python
import asyncio
# pykis 모듈 전체를 가져옵니다.
import pykis
import asyncio
from dotenv import load_dotenv
import os 

# 발급받은 App Key, App Secret 설정
load_dotenv()
APP_KEY = os.getenv('APP_KEY')
APP_SECRET = os.getenv('APP_SECRET')
ACCOUNT_NUMBER = os.getenv('ACCOUNT_NUMBER')


# 2. 실시간 데이터를 처리할 함수 정의
def on_price_update(data: pykis.KisRealtimePrice):
    """실시간 체결가 데이터가 들어올 때마다 이 함수가 실행됩니다."""
    print(f"[실시간] {data.stock_code}: 현재가 {int(data.price):,}원 | 등락률 {data.change_rate:.2f}%")

async def main():
    ws_client = None
    try:
        # 3. PyKis 객체 생성
        kis = pykis.PyKis(
            id=ACCOUNT_NUMBER,
            appkey=APP_KEY,
            secretkey=APP_SECRET,
            # 모의투자 계좌인 경우 virtual_account=True 추가
            # virtual_account=True
        )
        
        # 4. 웹소켓 클라이언트 객체 할당
        ws_client = kis.websocket

        # 5. 실시간 시세 구독 (가장 중요한 변경점)
        # 클래스와 종목코드를 별도 인자로 전달합니다.
        await ws_client.subscribe(pykis.KisRealtimePrice, "005930")
        await ws_client.subscribe(pykis.KisRealtimePrice, "000660")

        ws_client.add_handler(on_price_update)
        
        print("실시간 주가 정보 수신을 시작합니다. (60초 후 종료)")
        await asyncio.sleep(60)

    except Exception as e:
        print(f"오류 발생: {e}")
    finally:
        # 6. 모든 구독 취소
        if ws_client:
            await ws_client.unsubscribe_all()
            print("모든 실시간 데이터 구독을 취소하고 연결을 종료합니다.")


# Jupyter Notebook 등에서 실행
await main()
```
