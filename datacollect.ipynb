```python
import pandas as pd, yfinance as yf
start=str((pd.Timestamp.utcnow().normalize()-pd.Timedelta(days=365*3)).date())
end=str((pd.Timestamp.utcnow().normalize()-pd.Timedelta(days=1)).date())
yf.download(["SPY","QQQM","277630.KS","272910.KS","IMTB"], start=start, end=end, interval="1d", auto_adjust=True, threads=False).to_csv("prices_3y.csv")
```

```python
session = requests.Session(impersonate="chrome")

# 세션을 yfinance 객체에 직접 주입
ticker = yf.Ticker(["SPY","QQQM","277630.KS","272910.KS","IMTB"], session=session)
data = ticker.history(period="1d")

print(data)
```
